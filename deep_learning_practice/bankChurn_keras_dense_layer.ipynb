```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df.sample(5)
```

```python
df.drop('RowNumber', axis='columns',inplace=True)
df.drop('CustomerId', axis='columns',inplace=True)
df.drop('Surname', axis='columns',inplace=True)
```

```python
df.replace({'Female':1,'Male':0},inplace=True)
```

```python
df = pd.get_dummies(data=df,columns=['Geography'])
df.columns
```

```python
df['Geography_France'] = df['Geography_France'].astype(int)
```

```python
df['Geography_Germany'] = df['Geography_Germany'].astype(int)
df['Geography_Spain'] = df['Geography_Spain'].astype(int)
```

```python
df.sample(5)
```

```python
cols_to_scaling = ['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df[cols_to_scaling] = scaler.fit_transform(df[cols_to_scaling])
```

```python
df.sample(5)
```

```python
for col in df:
    print(f'{col}: {df[col].unique()}')
```

```python
X = df.drop('Exited',axis='columns')
y = df['Exited']

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2, random_state=5)
```

For Imbalance Data

```python
def ANN(X_train, y_train):
  model = keras.Sequential(
    [keras.layers.Dense(12,input_shape=(12,),activation='relu'),
     keras.layers.Dense(13,activation='relu'),
     keras.layers.Dense(1,activation='sigmoid')]
)

  model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
  model.fit(X_train, y_train, epochs=100)
  print('finish')
  yp = model.predict(X_test)
  print('finish 2',yp.shape)

  y_pred = np.round(yp)
  print(y_pred.shape, y_test.shape)
  from sklearn.metrics import confusion_matrix , classification_report

  print(classification_report(y_test,y_pred))
  return y_pred


```

```python
y_pred = ANN(X_train,y_train)
```

Undersample

```python
df_train = pd.concat([X_train, y_train], axis=1)
df_train.head()
```

```python
count_class_0, count_class_1 = df_train.Exited.value_counts()
count_class_0, count_class_1
```

```python
df_class_0 = df_train[df_train['Exited'] == 0]
df_class_1 = df_train[df_train['Exited'] == 1]
```

```python
df_class_0_under = df_class_0.sample(count_class_1)
df_test_under = pd.concat([df_class_0_under,df_class_1],axis=0)
df_test_under.Exited.value_counts()
```

```python
X_train_under = df_test_under.drop('Exited', axis='columns')
y_train_under = df_test_under['Exited']
y_pred = ANN(X_train_under,y_train_under)
```

```python
df_class_1_over = df_class_1.sample(count_class_0,replace=True)
df_test_over = pd.concat([df_class_0,df_class_1_over],axis=0)
```

```python
X_over = df_test_over.drop('Exited', axis='columns')
y_over = df_test_over['Exited']

y_pred = ANN(X_over,y_over)
```

SMOTE

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X_train, y_train)

y_sm.value_counts()
```

```python
y_pred = ANN(X_sm,y_sm)
```

```python
def getBatch(start,end):
  df_test_batch = pd.concat([df_class_0[start:end],df_class_1],axis=0)
  X_train_batch = df_test_batch.drop('Exited', axis='columns')
  y_train_batch = df_test_batch.Exited
  return X_train_batch, y_train_batch
```
