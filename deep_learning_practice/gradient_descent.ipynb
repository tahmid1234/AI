```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df[['age','affordibility']],df.bought_insurance,test_size=0.2, random_state=25)
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age'] / 100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age'] / 100
```

```python
model = keras.Sequential([keras.layers.Dense(1,input_shape=(2,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')])
model.compile(
    optimizer='adam',
     loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(X_train,y_train, epochs=8000)
```

```python
model.evaluate(X_test_scaled,y_test)
```

```python
def sigmoid_numpy(X):
   return 1/(1+np.exp(-X))
```

```python
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def gradient_descent(age, affordability, y_true, epochs, loss_thresold):
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age)
    for i in range(epochs):
        weighted_sum = w1 * age + w2 * affordability + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        loss = log_loss(y_true, y_predicted)

        w1d = (1/n)*np.dot(np.transpose(age),(y_predicted-y_true))
        w2d = (1/n)*np.dot(np.transpose(affordability),(y_predicted-y_true))

        bias_d = np.mean(y_predicted-y_true)
        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        bias = bias - rate * bias_d

        print (f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')

        if loss<=loss_thresold:
            break

    return w1, w2, bias
```

```python
class MyNN:
  def __init__(self):
    self.w1 = 0
    self.w2 = 0
    self.bias = 0

  def getWeightedSum(self,X1,X2,w1,w2,b):
    return w1*X1+w2*X2 + b

  def numpy_sigmoid(self,X):
   return 1/(1+np.exp(-X))

  def log_loss(self,y_true,y_pred):
    epsilon = 1e-15

    y_pred_new = [max(val,epsilon) for val in y_pred]
    y_pred_new = [min(val,1-epsilon) for val in y_pred]
    y_pred_new = np.array(y_pred_new)
    return - np.mean(y_true*np.log(y_pred_new)+(1-y_true)*np.log(1-y_pred_new))



  def predict (self, X):
    weighted_sum = self.getWeightedSum(X.age, X.affordibility,self.w1,self.w2,self.bias)
    return self.numpy_sigmoid(weighted_sum)



  def fit(self,X,y_true,epochs,loss_thresold):

    self.w1, self.w2, self.bias = self.gradient_descent(X,y_true,epochs,loss_thresold)
    print(f"Final weights and bias: w1: {self.w1}, w2: {self.w2}, bias: {self.bias}")



  def gradient_descent(self,X,y_true,epoch,loss_th):
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(X['age'])
    print(epoch,'epoch')
    for i in range (epoch):
      weighted_sum = self.getWeightedSum(X['age'],X['affordibility'],w1,w2,bias)
      y_pred = self.numpy_sigmoid(weighted_sum)
      loss = self.log_loss(y_true,y_pred)
      w1d = (1/n)*np.dot(np.transpose(X.age),(y_pred-y_true))
      w2d = (1/n)*np.dot(np.transpose(X['affordibility']),(y_pred-y_true))

      bias_d = np.mean(y_pred-y_true)

      w1 = w1- rate* w1d
      w2 = w2 - rate * w2d
      bias = bias - rate * bias_d

      if i%100==0:
        print (f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')

      if loss<= loss_th:
        print('loss',epoch,loss)
        break

    return w1, w2, bias



```
